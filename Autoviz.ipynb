```python
#Create Autoviz class
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

```python
import pandas as pd
import seaborn as sns
```

# Tips dataset

```python
#import tips dataset
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
tips_df.to_csv('tips_df.csv')
```

```python
#Done on tips data
filename = 'tips_df.csv'
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
